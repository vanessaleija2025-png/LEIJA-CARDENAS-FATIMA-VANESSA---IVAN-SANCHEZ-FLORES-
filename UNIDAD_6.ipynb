```python
#Ejercicio 1 — Resolver una EDO simple
import numpy as np

# Ecuación diferencial
# dy/dx = x - y

def f(x, y):
    return x - y

# Valores iniciales
x = 0
y = 1
h = 0.2

# Método de Euler
for i in range(10):
    y = y + h * f(x, y)
    x = x + h

    print("x =", round(x, 2), " y =", round(y, 4))
```

```python
#Ejercicio 2 — Visualización de la solución
import numpy as np
import matplotlib.pyplot as plt

# Ecuación diferencial
# dy/dx = x + y

def f(x, y):
    return x + y

# Valores iniciales
x = 0
y = 1
h = 0.1

# Listas para almacenar resultados
x_vals = []
y_vals = []

# Método de Euler
for i in range(20):
    x_vals.append(x)
    y_vals.append(y)

    y = y + h * f(x, y)
    x = x + h

# Gráfica
plt.plot(x_vals, y_vals, marker='o')
plt.title("Solución aproximada de la EDO")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

```python
#Ejercicio 3 — Aplicación IA: dinámica de aprendizaje
import matplotlib.pyplot as plt

# Simulación de aprendizaje
aprendizaje = 1.0
h = 0.1

valores = []

for i in range(30):
    aprendizaje = aprendizaje - h * aprendizaje
    valores.append(aprendizaje)

# Gráfica
plt.plot(valores)
plt.title("Dinámica de aprendizaje")
plt.xlabel("Iteraciones")
plt.ylabel("Valor")
plt.grid(True)
plt.show()
```

```python
#Ejercicio 2 — Comparación con solución exacta
import numpy as np
import matplotlib.pyplot as plt

# Ecuación diferencial
# dy/dx = y

def f(x, y):
    return y

# Valores iniciales
x = 0
y = 1
h = 0.1

x_vals = []
y_euler = []
y_exacta = []

# Método de Euler
for i in range(20):
    x_vals.append(x)
    y_euler.append(y)
    y_exacta.append(np.exp(x))

    y = y + h * f(x, y)
    x = x + h

# Gráfica
plt.plot(x_vals, y_euler, label="Euler")
plt.plot(x_vals, y_exacta, label="Exacta")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Ejercicio 3 — Aplicación IA: descenso dinámico
import matplotlib.pyplot as plt

# Descenso dinámico
peso = 10
h = 0.1

pesos = []

for i in range(50):
    gradiente = 2 * peso
    peso = peso - h * gradiente
    pesos.append(peso)

# Gráfica
plt.plot(pesos)
plt.title("Descenso dinámico")
plt.xlabel("Iteraciones")
plt.ylabel("Peso")
plt.grid(True)
plt.show()
```

```python
#Ejercicio 1 — Método RK2
import numpy as np

# Ecuación diferencial
# dy/dx = x + y

def f(x, y):
    return x + y

x = 0
y = 1
h = 0.1

for i in range(10):
    k1 = h * f(x, y)
    k2 = h * f(x + h, y + k1)

    y = y + (k1 + k2) / 2
    x = x + h

    print("x =", round(x, 2), " y =", round(y, 4))
```

```python
#Ejercicio 2 — Método RK4
import numpy as np

# Ecuación diferencial
# dy/dx = x + y

def f(x, y):
    return x + y

x = 0
y = 1
h = 0.1

for i in range(10):
    k1 = h * f(x, y)
    k2 = h * f(x + h/2, y + k1/2)
    k3 = h * f(x + h/2, y + k2/2)
    k4 = h * f(x + h, y + k3)

    y = y + (k1 + 2*k2 + 2*k3 + k4) / 6
    x = x + h

    print("x =", round(x, 2), " y =", round(y, 4))
```

```python
#Ejercicio 3 — Comparación Euler vs RK4
import numpy as np
import matplotlib.pyplot as plt

# Ecuación diferencial
# dy/dx = y

def f(x, y):
    return y

# Valores iniciales
x1 = 0
x2 = 0
y_euler = 1
y_rk4 = 1
h = 0.1

x_vals = []
euler_vals = []
rk4_vals = []

for i in range(20):
    x_vals.append(x1)
    euler_vals.append(y_euler)
    rk4_vals.append(y_rk4)

    # Euler
    y_euler = y_euler + h * f(x1, y_euler)
    x1 = x1 + h

    # RK4
    k1 = h * f(x2, y_rk4)
    k2 = h * f(x2 + h/2, y_rk4 + k1/2)
    k3 = h * f(x2 + h/2, y_rk4 + k2/2)
    k4 = h * f(x2 + h, y_rk4 + k3)

    y_rk4 = y_rk4 + (k1 + 2*k2 + 2*k3 + k4) / 6
    x2 = x2 + h

# Gráfica
plt.plot(x_vals, euler_vals, label="Euler")
plt.plot(x_vals, rk4_vals, label="RK4")
plt.legend()
plt.grid(True)
plt.show()
```
