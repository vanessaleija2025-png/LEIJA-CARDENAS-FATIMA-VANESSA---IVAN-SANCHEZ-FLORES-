```python
# Importa la librería numpy para operaciones numéricas.
import numpy as np

# Define el array de valores de entrada x.
x = np.array([1, 2, 3])
# Define el array de valores de salida y.
y = np.array([1, 4, 9])

# Obtiene el número de puntos de datos.
n = len(x)

# Inicializa una matriz (tabla) de ceros para almacenar las diferencias divididas.
tabla = np.zeros((n, n))

# Copia los valores de y en la primera columna de la tabla.
tabla[:, 0] = y

# Bucle anidado para calcular las diferencias divididas.
for j in range(1, n):
    for i in range(n - j):
        # Calcula la diferencia dividida actual.
        tabla[i, j] = (tabla[i+1, j-1] - tabla[i, j-1]) / (x[i+j] - x[i])

# Define la función `newton_eval` para evaluar el polinomio interpolador de Newton.
def newton_eval(x_data, tabla, x_val):
    # Obtiene el número de puntos de datos.
    n = len(x_data)
    # Inicializa el resultado con el primer término de la tabla.
    resultado = tabla[0, 0]

    # Inicializa el producto para los términos (x_val - x_i).
    producto = 1

    # Itera para sumar los términos restantes del polinomio de Newton.
    for i in range(1, n):
        # Actualiza el producto acumulativo.
        producto *= (x_val - x_data[i-1])
        # Suma el término actual al resultado.
        resultado += tabla[0, i] * producto

    # Retorna el valor interpolado.
    return resultado

# Llama a la función `newton_eval` para interpolar en x_val = 2.5.
valor = newton_eval(x, tabla, 2.5)

# Imprime el valor interpolado.
print("Valor interpolado:", valor)
```

```python
# Importa la librería numpy para operaciones numéricas.
import numpy as np

# Define el array de valores de entrada x.
x = np.array([1,2,3])
# Define el array de valores de salida y.
y = np.array([1,4,9])

# Define la función `lagrange` para realizar la interpolación de Lagrange.
def lagrange(x_data, y_data, x_val):
    # Obtiene el número de puntos de datos.
    n = len(x_data)
    # Inicializa el resultado del polinomio interpolador.
    resultado = 0

    # Bucle principal para sumar los términos del polinomio de Lagrange.
    for i in range(n):
        # Inicializa el término L_i(x) (polinomio base de Lagrange).
        L = 1

        # Bucle para calcular el producto dentro de cada término L_i(x).
        for j in range(n):
            # Excluye el caso cuando i es igual a j.
            if j != i:
                # Multiplica los factores (x_val - x_j) / (x_i - x_j).
                L *= (x_val - x_data[j]) / (x_data[i] - x_data[j])

        # Suma el término y_i * L_i(x) al resultado total.
        resultado += y_data[i] * L

    # Retorna el valor interpolado.
    return resultado

# Llama a la función `lagrange` para interpolar en x_val = 2.5.
valor = lagrange(x, y, 2.5)
# Imprime el valor interpolado.
print("Valor interpolado:", valor)
```

```python
# Importa la librería numpy para operaciones numéricas.
import numpy as np

# Define una función simple f(x) = x^2.
def f(x):
  return x**2

# Define el array de valores de entrada x.
x = np.array([1,2,3])
# Calcula los valores de y aplicando la función f a cada elemento de x.
y = f(x)

# Define la función `lagrange` para realizar la interpolación de Lagrange.
def lagrange(x_data,y_data,x_val):
  # Obtiene el número de puntos de datos.
  n = len(x_data)
  # Inicializa el resultado del polinomio interpolador.
  resultado = 0

  # Bucle principal para sumar los términos del polinomio de Lagrange.
  for i in range(n):
        # Inicializa el término L_i(x).
        L = 1

        # Bucle para calcular el producto dentro de cada término L_i(x).
        for j in range(n):
            # Excluye el caso cuando i es igual a j.
            if j != i:
                # Multiplica los factores (x_val - x_j) / (x_i - x_j).
                L *= (x_val - x_data[j]) / (x_data[i] - x_data[j])

        # Suma el término y_i * L_i(x) al resultado total.
        resultado += y_data[i] * L

  # Retorna el valor interpolado.
  return resultado

# Define el punto en el que se desea interpolar.
x_test = 2.5

# Calcula el valor interpolado usando la función de Lagrange.
interp = lagrange(x,y,x_test)

# Calcula el valor real de la función en x_test.
real = f(x_test)

# Calcula el error absoluto entre el valor real y el interpolado.
error = abs(real - interp)

# Imprime el valor interpolado.
print("interpolado:",interp)
# Imprime el valor real.
print("real:",real)
# Imprime el error absoluto.
print("error:",error)
```

```python
# Importa la librería numpy para operaciones numéricas.
import numpy as np
# Importa CubicSpline de scipy.interpolate para interpolación por splines cúbicos.
from scipy.interpolate import CubicSpline

# Define el array de valores de entrada x.
x = np.array([0,1,2,3])
# Define el array de valores de salida y.
y = np.array([1,3,2,5])

# Crea un objeto CubicSpline usando los puntos de datos x e y.
spline = CubicSpline(x,y)

# Genera un array de 100 puntos igualmente espaciados entre 0 y 3 para la interpolación.
x_interp = np.linspace(0,3,100)
# Evalúa el spline cúbico en los nuevos puntos x_interp para obtener los valores interpolados y_interp.
y_interp = spline(x_interp)

# Imprime los valores interpolados.
print("valores interpolados:",y_interp)
```

```python
# Importa la librería numpy para operaciones numéricas.
import numpy as np

# Define el array de valores de entrada x.
x = np.array([1,2,3,4,5])
# Define el array de valores de salida y.
y = np.array([2,4,5,4,5])

# Realiza un ajuste polinómico de grado 1 (regresión lineal) a los datos x e y.
# 'coef' contendrá la pendiente (m) y el intercepto (b).
coef = np.polyfit(x,y,1)

# Extrae la pendiente (m) del resultado.
m = coef[0]
# Extrae el intercepto (b) del resultado.
b = coef[1]

# Calcula los valores de y predichos usando la ecuación de la línea (y = mx + b).
y_pred = m*x + b

# Imprime la pendiente.
print("pendiente:",m)
# Imprime el intercepto.
print("intercepto:",b)
# Imprime los valores predichos de y.
print("prediccion:",y_pred)
```
