```python
import numpy as np
import matplotlib.pyplot as plt

n = 120
x = np.linspace(0, 100, n)
y = 2.0 * x + np.random.normal(0, 25, n)

plt.scatter(x, y, color='steelblue', alpha=0.7)
plt.title('Relación entre X e Y')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
#Esta la escogimos por la grafica de burbuja ya que, queremos ver la relacion que hay entre X y Y y con un scatter tenemos la distancia de cada punto
```

```python
import matplotlib.pyplot as plt
import numpy as np

muestra = np.random.normal(loc=50, scale=10, size=1000) + np.random.exponential(scale=2, size=1000)*0.5

plt.hist(muestra, bins=30, color='skyblue', edgecolor='black', density=True)
plt.title('Distribución de la muestra')
plt.xlabel('Valor')
plt.ylabel('Densidad')
plt.show()
#Escogimos esta grafica ya que ayuda a visualizar la distribucion de una muestra y nos indica la densidad de cada una de ellas
```

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(1, 25)
tendencia = 5 + 1.2 * t
estacional = 4 * np.sin(2 * np.pi * t / 12)
ruido = np.random.normal(0, 2, size=t.shape)
serie = tendencia + estacional + ruido

plt.plot(t, serie, marker='o', color='orange')
plt.title('Serie temporal con tendencia y estacionalidad')
plt.xlabel('Tiempo (t)')
plt.ylabel('Valor de la serie')
plt.grid(True)
plt.show()
#escogimos esta grafica nos ayuda a ver que tendencia tiene una serie y el comportamiento temporal
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Re-defining df to make this cell self-contained
df = pd.DataFrame({
    'ingresos': np.random.normal(50, 10, 200),
    'gasto_marketing': np.random.normal(25, 5, 200),
    'clientes_nuevos': np.random.poisson(12, 200) + np.random.normal(0, 1, 200),
    'satisfaccion': np.clip(np.random.normal(3.5, 0.7, 200), 1, 5)
})
# Inducimos una variable dependiente combinada (ventas).
df['ventas'] = 2.5*df['ingresos'] + 1.1*df['gasto_marketing'] + np.random.normal(0, 8, 200)

plt.scatter(df['ingresos'], df['ventas'], color='steelblue', alpha=0.7)
plt.title('Relación entre ingresos y ventas')
plt.xlabel('Ingresos')
plt.ylabel('Ventas')
plt.show()
#escogimos otra vez esta grafica de burbuja ya que tambien tiene relacion en ventas y los ingresos.
```

```python
import matplotlib.pyplot as plt
import numpy as np

fs = 500
T  = 2.0
t = np.linspace(0, T, int(fs*T), endpoint=False)
f1, f2 = 50, 120
signal = 1.0*np.sin(2*np.pi*f1*t) + 0.6*np.sin(2*np.pi*f2*t) + 0.3*np.random.normal(0,1,len(t))

plt.figure(figsize=(10,4))
plt.plot(t, signal, color='steelblue')
plt.title('Señal en el dominio del tiempo')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.grid(True)
plt.show()
#escogimos esta grafica para visualizar alguna señal en el dominio del tiempo con la amplitud de cada dato.
```
