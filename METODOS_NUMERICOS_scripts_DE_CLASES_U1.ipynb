```python
# Aproximación numérica de la raíz de f(x) = x^2 - 2

import numpy as np  # Librería para operaciones numéricas

# Genera 1000 puntos entre 0 y 2
x = np.linspace(0, 2, 1000)

# Evalúa la función
f = x**2 - 2

# Encuentra el punto donde f(x) está más cerca de 0
x_aprox = x[np.argmin(np.abs(f))]

# Muestra la raíz aproximada
print("Raíz aproximada:", x_aprox)
```

```python
# Ejemplo de redondeo de un número decimal

import numpy as np  # Se importa numpy aunque no es necesario para usar round

x = 3.1415926535  # Se define un número decimal

# round redondea el número a la cantidad de decimales indicada

print(round(x, 2))  # Redondea a 2 decimales
print(round(x, 4))  # Redondea a 4 decimales
print(round(x, 6))  # Redondea a 6 decimales
```

```python
# Cálculo de media y error absoluto

import numpy as np  # Librería para operaciones numéricas

# Arreglo con las mediciones obtenidas
mediciones = np.array([9.8, 9.81, 9.79, 9.80])

# Valor real o de referencia
valor_real = 9.81

# Calcula la media de las mediciones
media = np.mean(mediciones)

# Calcula el error absoluto
# Es la diferencia entre la media y el valor real
error = abs(media - valor_real)

# Muestra los resultados
print("media:", media)
print("error absoluto:", error)
```

```python
# Ejemplo de sesgo en un conjunto de datos

import numpy as np  # Librería para generar datos y hacer cálculos

# Genera 1000 datos aleatorios con distribución normal
# Media 0 y desviación estándar 1
datos_reales = np.random.normal(0, 1, 1000)

# Se define un sesgo constante
sesgo = 0.5

# Se agregan 0.5 a todos los datos
# Esto desplaza la media hacia arriba
datos_sesgados = datos_reales + sesgo

# Se calcula y muestra la media de ambos conjuntos
print("Media real:", np.mean(datos_reales))
print("Media sesgada:", np.mean(datos_sesgados))
```

```python
# Ejemplo de error por representación en punto flotante

# Se suman dos números decimales
a = 0.1 + 0.2

# Se define otro número decimal
b = 0.3

# Muestra el resultado de la suma
print(a)

# Compara si ambos valores son exactamente iguales
print(a == b)

# Calcula la diferencia absoluta entre ambos
print(abs(a - b))
```

```python
# Aproximación de e^x usando serie de Taylor hasta segundo orden

import math  # Librería para funciones matemáticas

x = 1  # Valor en el que se evaluará la función

# Aproximación de e^x ≈ 1 + x + x^2/2
aprox = 1 + x + (x**2) / 2

# Valor real usando la función exponencial
real = math.exp(x)

# Cálculo del error absoluto
error = abs(real - aprox)

# Mostrar resultados
print("aprox:", aprox)
print("real:", real)
print("error:", error)
```

```python
# Cálculo de error propagado usando derivadas

x = 2        # Valor de la variable
dx = 0.01    # Pequeña variación o incertidumbre en x

# Función y = x^2
y = x**2

# La derivada de f(x) = x^2 es f'(x) = 2x
# El error propagado se aproxima como |f'(x)| * dx
dy = abs(2 * x) * dx

# Mostrar resultados
print("y:", y)
print("error aproximado en y (dy):", dy)
```

```python
import numpy as np

# Definimos la matriz A con corchetes anidados para 2D
A = np.array([[3, 2],
              [1, 2]])

b = np.array([5, 5])

# Resolvemos el sistema de ecuaciones Ax = b
x = np.linalg.solve(A, b)
print(x)
```

```python
# Minimización de una función usando scipy

from scipy.optimize import minimize  # Función para encontrar mínimos

# Se define la función f(x) = (x - 2)^2
# Esta función tiene un mínimo en x = 2
f = lambda x: (x - 2)**2

# Se aplica el método de minimización
# x0=0 es el valor inicial desde donde comienza la búsqueda
resultado = minimize(f, x0=0)

# Muestra el valor de x donde la función alcanza su mínimo
print("Solución (x):", resultado.x)

# Muestra toda la información del proceso de optimización
print(resultado)
```
