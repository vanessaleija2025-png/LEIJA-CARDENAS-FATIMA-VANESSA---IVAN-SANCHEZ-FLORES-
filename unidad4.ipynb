```python
import numpy as np

# Define la función f(x) = x^3
def f(x):
  return x**3

x = 2 # Punto en el que se calculará la derivada

h = 0.001 # Pequeño tamaño de paso para la aproximación numérica

# Calcula la derivada numérica utilizando la fórmula de diferencia hacia adelante
derivada = (f(x+h) - f(x)) / h

# Calcula la derivada real (analítica): d/dx(x^3) = 3*x^2
real = 3*x**2

# Calcula el error absoluto entre las derivadas numérica y real
error = abs(real - derivada)

# Imprime los resultados
print("Derivada numerica:", derivada)
print("Derivada real:",real)
print("Error:",error)
```

```python
import numpy as np

# Define la función f(x) = x^3
def f(x):
  return x**3

x = 2 # Punto en el que se calculará la derivada

h=0.001 # Pequeño tamaño de paso para la aproximación numérica

# Calcula la derivada numérica utilizando la fórmula de diferencia central
derivada = (f(x+h)-f(x-h))/(2*h)

# Calcula la derivada real (analítica): d/dx(x^3) = 3*x^2
real = 3*x**2

# Calcula el error absoluto entre las derivadas numérica y real
error = abs(real-derivada)

# Imprime los resultados
print("Derivada central:", derivada)
print("Derivada real:",real)
print("Error:",error)
```

```python
# aproximacion del gradiente (Gradient Approximation)

import numpy as np

# Define la función de pérdida loss(w) = w^2 + 3w + 1
def loss(w):
  return w**2 + 3*w + 1

w = 1.5 # Punto en el que se aproximará el gradiente
h = 0.0001 # Pequeño tamaño de paso para la aproximación numérica

# Calcula el gradiente numérico utilizando la fórmula de diferencia central
grad = (loss(w+h)-loss(w-h))/(2*h)

# Calcula el gradiente real (analítico): d/dw(w^2 + 3w + 1) = 2w + 3
real = 2*w + 3

# Imprime los resultados
print("Gradiente numerico:",grad)
print("Gradiente real:",real)
```

```python
# trapecio (Trapezoidal Rule for Integration)
import numpy as np

# Define la función f(x) = x^2
def f(x):
    return x**2

a = 0 # Límite inferior de integración
b = 2 # Límite superior de integración
n = 100 # Número de subintervalos

h = (b - a) / n # Ancho de cada subintervalo

# Genera n+1 puntos entre a y b, ambos inclusive
x = np.linspace(a, b, n+1)

# Evalúa la función en estos puntos
y = f(x)

# Aplica la fórmula de la Regla del Trapecio:
# Integral = (h/2) * [f(x0) + 2*sum(f(xi) for i=1 to n-1) + f(xn)]
integral = (h/2) * (y[0] + 2*np.sum(y[1:n]) + y[n])

# Imprime la integral aproximada
print ("Integral aproximada (Trapecio):", integral)
```

```python
# simpson (Simpson's Rule for Integration)
# La función f(x) = x^2 está definida en la celda anterior.

a = 0 # Límite inferior de integración
b = 2 # Límite superior de integración
n = 100 # Número de subintervalos (debe ser par para la Regla de Simpson)

h = (b-a)/n # Ancho de cada subintervalo

# Genera n+1 puntos entre a y b, ambos inclusive
x = np.linspace(a,b,n+1)

# Evalúa la función en estos puntos (se asume f(x) de la celda anterior)
y = f(x)

# Inicializa la suma para la Regla de Simpson
suma = y[0] + y[n] # Añade el primer y último término

# Itera a través de los puntos intermedios, aplicando pesos (4 para índices impares, 2 para índices pares)
for i in range(1,n):
  if i % 2 == 0: # Índice par
    suma += 2*y[i]
  else: # Índice impar
    suma += 4*y[i]

# Aplica la fórmula de la Regla de Simpson: Integral = (h/3) * suma
integral = (h/3)*suma

# Imprime la integral aproximada
print("integral aproximada (simpson):",integral)
```
