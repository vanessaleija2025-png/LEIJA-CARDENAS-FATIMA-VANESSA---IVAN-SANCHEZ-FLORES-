```python
import pandas as np  # Importa la librería (aunque normalmente se usa numpy)

# Definición de la función f(x)
def f(x):
    return x**3 - x - 2

# Valores iniciales del intervalo
a = 1
b = 2

# Evaluación de la función en los extremos del intervalo
fa = f(a)  # valor de la función en a
fb = f(b)  # valor de la función en b

# Mostrar resultados en pantalla
print("f(a):", fa)
print("f(b):", fb)
```

```python
# Verifica si hay cambio de signo entre f(a) y f(b)
if fa * fb < 0:  # condición del teorema del valor intermedio
    print("existe al menos una raiz en el intervalo")  # confirma que hay una raíz
else:
    print("no existe raiz en el intervalo")  # no hay cambio de signo
```

implementar biseccion manual

```python
# Definición de la función
def f(x):
    return x**3 - x - 2  # función no lineal

# Intervalo inicial
a = 1  # extremo izquierdo
b = 2  # extremo derecho

tolerancia = 1e-6  # error máximo permitido
max_iter = 23  # número máximo de iteraciones

# Ciclo del método de bisección
for i in range(max_iter):
    c = (a + b) / 2  # cálculo del punto medio
    fc = f(c)  # evaluación de la función en el punto medio

    # Criterio de paro por tolerancia
    if abs(fc) < tolerancia:
        break

    # Verifica en qué subintervalo está la raíz
    if f(a) * fc < 0:
        b = c  # la raíz está entre a y c
    else:
        a = c  # la raíz está entre c y b

# Resultados
print("Raiz aproximada:", c)
print("Iteraciones realizadas:", i+1)
```

```python
a = 1  # límite inferior del intervalo
b = 2  # límite superior del intervalo
n = 22  # número de iteraciones realizadas

# Cálculo del error máximo teórico en el método de bisección
error_teorico = (b - a) / (2**n)

# Mostrar el error máximo estimado
print("error maximo despues de", n, "iteraciones:", error_teorico)
```

```python
import numpy as np # biblioteca para calculos numericos

def g(x): # definicion de la funcion iterativa  g(x)
    return np.cos(x) #funcion de coseno que define el punto fijo

x = 0.5 # valor inicial de la aproximacion
tolerancia = 1e-6 # error maximo permitido
max_iter = 100 # numero maximo de iteraciones

for i in range(max_iter):# ciclo de iteraciones
    x_new = g(x) # aplicacion de la formula de aproximacion susesiva

    if abs(x_new - x)< tolerancia: # criterio de paro basado en el erro absoluto
        break # termina si la diferencia es suficientemente pequeña

    x = x_new # actualizada del valor para la siguiennte iteracion

print("raiz aproximada :",x_new)# mostrar resultado final
print("iteracionees realizadas :",i+1)#mostrar numero de iteraciones
```

interpolacion lineal manual (1,2)(3,6)

```python
x0 = 1 #primer punto en x
y0 = 2 #valor correspondiente en y
x1 = 3 #segundo punto en x
y1 = 6 #valor correspondiente en y

x = 2 #punto donde se desea interpolacion

p = y0 + ((y1 - y0) / (x1 - x0)) * (x - x0) #formula

print("valor interpolado:", p) #mostrar resultado
```

```python
import numpy as np #Biblioteca numerica

x_datos = np.array([0,1,2]) #Arreglo con valores conocidos en x
y_datos = np.array([1,3,2]) #Arreglo con valores conocidos en y

coeficientes = np.polyfit(x_datos, y_datos, 2)
polinomio = np.poly1d(coeficientes) #construccion

x_eveal = 1.5 #punto donde se evaluara el polinomio
y_eveal = polinomio(x_eveal) #evaluacion del polinomio

print("Coeficientes del polinomio: ", coeficientes)
print("Valor interpolado:", y_eveal ) #mostrar valor
```
