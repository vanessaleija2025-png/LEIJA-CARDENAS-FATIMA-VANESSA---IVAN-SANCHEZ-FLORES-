# Ejercicio 3 - Metodo de Gauss-Seidel

```python
import numpy as np

x = np.zeros(3)

tol = 1e-6

for k in range(100):

  x_old = x.copy()

  x[0] = (11 - x[1] + x[2]) / 10

  x[1] = (12 - 2*x[0] - x[2]) / 10

  x[2] = (13 - 2*x[0] - 3*x[1]) / 10

  if np.linalg.norm(x - x_old) < tol:
    break

print("Solucion Gauss-Seidel : ",x)
print("iteraciones : ",k+1)
```

# Ejercicio 2 - Resolución con Scipy

```python
import numpy as np #Blibloteca para calculos numericos
from scipy.optimize import fsolve # importa el metodo fsolve para resolver no lineales

def sistema(vars): # Define la funcion que representa el sistema de ecuaciones
  x,y = vars # Separa las variables del vector de entrada
  f1 = x**2 + y**2 - 4 # primera ecuacion del sistema
  f2 = x - y -1 # segunda ecuacion del sistema
  return [f1,f2] # devuelve el sistema evaluado

sol = fsolve(sistema,[1,1]) # busca la solucion usando como aproximacion inicial (1,1)

print("Solucion aproximada : ",sol) # imprime la solucion encontrada
```

# Ejercicio 1 - Verificacion de Convergencia

```python
import numpy as np

A = np.array([[10,1,-1],[2,10,1],[2,3,10]]) #Define la primera fila de la matriz de coeficientes del sistema de ecuaciones

D = np.diag(np.diag(A)) # Extrae la matriz diagonal A

L = -np.tril(A,-1) #Obtiene la parte triangular inferior de la matriz A con signo

U = -np.triu(A,1)  #Obtiene la parte triangular inferior de la matriz A con signo

T = np.linalg.inv(D).dot(L+U) # Calcula la matriz de iteracion del metodo del metodo de jacobi

eligenvalues = np.linalg.eigvals(T) # calcula los eigenvalores de la matriz de iteraciones

rho = max(abs(eligenvalues)) # calcula los eigenvalores de la matriz

print("Radio espectral : ", rho) # imprime el valor del radio espectral

if rho < 1: # verifica si el radio espectral es menor que uno
  print("El metodo converge") # indica que el metodo iterativo converge

else:
  print("El metodo  no converge")# indica que el metodo iterativo no converge
```

# **Ejercicio 3 - Aplicacion: Regresion Lineal**

```python
import numpy as np

X = np.array([[1,1],[1,2],[1,3],[1,4]]) # matriz de datos donde la primera columna es el termino independiente
#segunda observacion del conjunto de datos , Tercera observacion , Cuarta observacion

y = np.array([6,5,7,10]) # Vector de valores observados

XT = X.T # Calcula la transpuesta de la matriz de datos

beta = np.linalg.inv(XT.dot(X)).dot(XT).dot(y)#calcula los coeficientes de la regresion lineal

print("Coeficientes del modelo : ",beta)#imprime los parametros estimados del modelo
```
